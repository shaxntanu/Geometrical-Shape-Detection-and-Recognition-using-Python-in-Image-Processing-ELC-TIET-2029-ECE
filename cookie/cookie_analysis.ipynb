```python
# Import libraries\nimport cv2\nimport numpy as np\nfrom matplotlib import pyplot as plt\nfrom PIL import Image
```

```python
# Load and preprocess image\nimage_path = 'cookie.jpg'\nimg = Image.open(image_path)\nimg_array = np.array(img)\n\nimg_rgb = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)\ngray = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)\n\nplt.figure(figsize=(10, 5))\nplt.subplot(1, 2, 1)\nplt.imshow(img_rgb)\nplt.title('Original Cookie')\nplt.axis('off')\n\nplt.subplot(1, 2, 2)\nplt.imshow(gray, cmap='gray')\nplt.title('Grayscale')\nplt.axis('off')\nplt.show()
```

```python
# Detect circular cookie shape\nblurred = cv2.GaussianBlur(gray, (9, 9), 0)\ncircles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT, dp=1, minDist=100,\n                          param1=50, param2=30, minRadius=50, maxRadius=300)\n\noutput_image = img_array.copy()\n\nif circles is not None:\n    circles = np.uint16(np.around(circles))\n    print(f\"Detected {len(circles[0])} circular cookie(s)\")\n    \n    for i, circle in enumerate(circles[0, :]):\n        center = (circle[0], circle[1])\n        radius = circle[2]\n        \n        cv2.circle(output_image, center, radius, (0, 255, 0), 4)\n        cv2.circle(output_image, center, 2, (255, 0, 0), 5)\n        \n        print(f\"Cookie {i+1}: Center=({circle[0]}, {circle[1]}), Radius={radius}\")\n        \n        area = np.pi * (radius ** 2)\n        print(f\"Cookie area: {area:.2f} pixels²\")\n        \n        cv2.putText(output_image, \"Cookie\", (center[0]-40, center[1]), \n                   cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 0, 0), 2)
```

```python
# Edge detection and thresholding\nedges = cv2.Canny(blurred, 30, 100)\n_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)\n\nplt.figure(figsize=(10, 5))\nplt.subplot(1, 2, 1)\nplt.imshow(edges, cmap='gray')\nplt.title('Edge Detection')\nplt.axis('off')\n\nplt.subplot(1, 2, 2)\nplt.imshow(thresh, cmap='gray')\nplt.title('Binary Threshold')\nplt.axis('off')\nplt.show()
```

```python
# Display final result\noutput_rgb = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)\nplt.figure(figsize=(8, 8))\nplt.imshow(output_rgb)\nplt.title('Detected Cookie Shape')\nplt.axis('off')\nplt.show()\n\nprint(\"\\nTexture Analysis:\")\nprint(f\"Mean intensity: {np.mean(gray):.2f}\")\nprint(f\"Standard deviation: {np.std(gray):.2f}\")
```
